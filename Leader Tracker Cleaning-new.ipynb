```python
import numpy as np
import pandas as pd
pd.set_option("display.max_rows", 60)
from datetime import date
print('Last run on: ' + str(date.today()))
```

```python
# Read all tabs you want from your Google spreadsheet and write in df
data = pd.ExcelFile('OOC 2021 Leaders.xlsx') 
df = pd.DataFrame() #where the imported data will go after formatting

for i in range(3, len(data.sheet_names)-1): #don't need first 3 or last tabs in file
    df_temp = pd.read_excel ('OOC 2021 Leaders.xlsx', sheet_name = data.sheet_names[i], header = 3)#Read an organizers sheet
    df_temp = df_temp.loc[:,:'Notes & Reflections'] #Take first column through notes column only
    df_temp['Organizer'] = data.sheet_names[i]#add column for organizer
    df_temp = df_temp.loc[~(df_temp['First Name'].isnull() & df_temp['Last Name'].isnull() & df_temp['E-Mail'].isnull()),:] #Remove rows with null first/last name/Email
    df_temp = df_temp.iloc[:-1] #Remove last row
    df = pd.concat([df, df_temp])
    
df.reset_index()

df = df.loc[~df['E-Mail'].str.contains('@ohorganizing.org', na=False), :] #Remove @ohorganizing.org emails  
```

```python
#overall counts
df.describe(include='all')
```

```python
#Column counts by organizer
df.groupby('Organizer')[['First Name', 'Last Name', 'E-Mail', 'Phone', 'Zip Code', 'Institution', 'Leadership Status', 'Recruited By']].count()
```

```python
# Output your df as an Excel document
pd_writer = pd.ExcelWriter('OOC 2021 Leaders Cleaned.xlsx')
df.to_excel(pd_writer, index=False, index_label=None, sheet_name="Master Sheet")
pd_writer.save()
pd_writer.close()
```
